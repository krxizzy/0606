```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


file_text = ""
for filename in ["/content/drive/MyDrive/colabtxt/how did it end lyrics.txt",
                 "/content/drive/MyDrive/colabtxt/fortnight.txt",
                 "/content/drive/MyDrive/colabtxt/the tourtured poets.txt"]:
    with open(filename, "r") as f:
        file_text += f.read()

wordcloud = WordCloud(
    background_color="black",
    width=2000,
    height=1000,
    margin=10,
).generate(file_text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
# Read the text
f = open('/content/drive/MyDrive/colabtxt/fortnight.txt', 'r').read()

# Stopwords (stop words)
stop = ["were", "there", "the", "when"] + list(STOPWORDS)

wordcloud = WordCloud(
    background_color="white",
    width=1500,
    height=1000,
    margin=2,
    stopwords=stop
).generate(f)


plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
